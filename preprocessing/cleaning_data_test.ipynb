```python
import json
from unicodedata import numeric
import pandas as pd


class Cleaning():
    
    df = pd.read_json("informations.json")
        
    def __init__(self):
        self.number_of_bedrooms = self.df.loc[["rooms-number"][0]][0]
        self.surface = self.df.loc[["area"][0]][0]
        self.fully_equipped_kitchen = self.df.loc[["equipped-kitchen"][0]][0]
        self.open_fire = self.df.loc[["open-fire"][0]][0]   
        self.number_of_facades = self.df.loc[["facades-number"][0]][0]
        self.terrace = self.df.loc[["terrace"][0]][0]
        self.terrace_surface = self.df.loc[["terrace-area"][0]][0]
        self.swimming_pool = self.df.loc[["swimming-pool"][0]][0]
        self.state_of_the_building = self.df.loc[["building-state"][0]][0]
        self.zip_code_ratio = self.df.loc[["zip-code"][0]][0]
        
    def check(self):
        # print(type(self.number_of_bedrooms))
        try:
            self.number_of_bedrooms = int(self.number_of_bedrooms)
        except ValueError:
            print('The provided value for "rooms-number" is not an integer')                 
        
        # print(type(self.surface))
        try:
            self.surface = int(self.surface)
        except ValueError:
            print('The provided value for "area" is not an integer')         
        
        # print(type(self.fully_equipped_kitchen))
        if self.fully_equipped_kitchen != 0 and self.fully_equipped_kitchen != 1:
            print('"equipped-kitchen" input error. Need to be 1 (Yes) or 0 (No).')        
        
        # print(type(self.open_fire))    
        if self.open_fire != 0 and self.open_fire != 1:
            print('"open-fire" input error. Need to be 1 (Yes) or 0 (No).')        
        
        # print(type(self.number_of_facades))
        try:
            self.number_of_facades = int(self.number_of_facades)
        except ValueError:
            print('The provided value for "facade-number" is not an integer')         
    
        # print(type(self.terrace))
        if self.terrace != 0 and self.terrace != 1:
            print('"terrace" input error. Need to be 1 (Yes) or 0 (No).')
        
    
        # print(type(self.terrace_surface))
        try:
            self.terrace_surface = int(self.terrace_surface)
        except ValueError:
            print('The provided value for "terrace-area" is not an integer')             
        
        
        # print(type(self.swimming_pool))
        if self.swimming_pool != 0 and self.swimming_pool != 1:
            print('"swimming-pool" input error. Need to be 1 (Yes) or 0 (No).')
        
        
        # print(type(self.state_of_the_building))
        if self.state_of_the_building == "NEW":
            self.state_of_the_building = int(1)
        elif self.state_of_the_building == "JUST RENOVATED" :
            self.state_of_the_building = int(0.75)
        elif self.state_of_the_building == "GOOD":
            self.state_of_the_building = int(0.5)
        elif self.state_of_the_building == "TO RENOVATE" or "TO REBUILD":
            self.state_of_the_building = int(0.25)
        else:
            print('"building-state" input error. The input need to be : "NEW" | "GOOD" | "TO RENOVATE" | "JUST RENOVATED" | "TO REBUILD"')

        # print(type(self.zip_code_ratio))
        if len(str(self.zip_code_ratio)) != 4:
            print('"zip-code" input error. Zip code need to contain 4 figures.')
        else:
            self.zip_code_ratio = 'be_zip_'+str(self.zip_code_ratio)[:-1]
            df_zip_code = pd.read_csv("preprocessing/diczipcoderatio.csv", names=["zipcode", "score"], skipinitialspace=True)
            self.zip_code_ratio = (df_zip_code["zipcode"] == self.zip_code_ratio)
            self.zip_code_ratio = df_zip_code[self.zip_code_ratio]["score"].values[0]
            print(self.zip_code_ratio)

        
        
        
        return        

        
```
